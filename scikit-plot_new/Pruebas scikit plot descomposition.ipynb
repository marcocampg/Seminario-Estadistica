```python
%load_ext autoreload
%autoreload 2

from scikitplot import decomposition
from sklearn.decomposition import PCA
from sklearn.datasets import load_digits, load_breast_cancer
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
Xx, yy = load_digits(return_X_y=True)
X_trainn, X_testt, y_trainn, y_testt = train_test_split(Xx, yy)
X, y = load_breast_cancer(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
pca = PCA()
pca.fit(X)
decomposition.plot_pca_component_variance(pca)
plt.show()
```

```python
decomposition.plot_pca_2d_projection(pca, X, y)
```
